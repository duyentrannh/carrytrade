```python

# Given values
price_per_unit_usd = 1000  # The price per unit in USD
total_units = 10000        # The total quantity of phones
cost_per_unit_nok = 7000   # The cost per unit in NOK
exchange_rate = 8  # NOK/USD meaning that it needs 8 NOK per one USD(The spot exchange rate NOK/USD)


#Define a function to compute the profit margin

def calculate_profit_margin(price_per_unit_usd, total_units, cost_per_unit_nok, exchange_rate):
    # Calculate total revenues in USD
    total_revenues_usd = price_per_unit_usd * total_units
    
    # Convert total revenues to NOK
    total_revenues_nok = total_revenues_usd * exchange_rate
    
    # Calculate total costs in NOK
    total_costs_nok = cost_per_unit_nok * total_units
    
    # Calculate total profit in NOK
    total_profit_nok = total_revenues_nok - total_costs_nok
    
    # Calculate profit per unit in NOK
    profit_per_unit_nok = total_profit_nok / total_units
    
    # Calculate price per unit in NOK
    price_per_unit_nok = price_per_unit_usd * exchange_rate
    
    # Calculate the profit margin
    profit_margin = profit_per_unit_nok / price_per_unit_nok
    
    print(f"Total revenues in USD: {total_revenues_usd:,} USD")
    print(f"Total revenues in NOK: {total_revenues_nok:,} NOK")
    print(f"Total costs in NOK: {total_costs_nok:,} NOK")
    print(f"Total profit in NOK: {total_profit_nok:,} NOK")
    print(f"Profit per unit in NOK: {profit_per_unit_nok:,} NOK")
    print(f"Price per unit in NOK: {price_per_unit_nok:,} NOK")
    return profit_margin

       
# Calculate the profit margin
profit_margin = calculate_profit_margin(price_per_unit_usd, total_units, cost_per_unit_nok, exchange_rate)

# Print the profit margin
print(f"The profit margin is: {profit_margin:,.2%}")
```

# A4
The financial deparment came up with such estimate of the exchange rate in year t + 1, using the concept of relative purchasing power parity(RPPP). They believe that RPPP will hold.

$ RPPP_t = S_{t-1} * \frac {1 +inflation_t} {1 + inflation_t^*} $

The results below show that the RPPP is the same as expected exchange rate. Hence ,RPPP shows that countries with higher inflation will experience currency depreciation, thus leaving the real exchange rate constant. 

```python

# Given data
exchange_rate = 8.0  # NOK/USD
inflation_norway = 0.01   # Percentage
inflation_usa = 0.05     # Percentage

def calculate_rppp_exchange_rate(exchange_rate, inflation_norway, inflation_usa):
    """
    Calculate the new exchange rate based on Relative Purchasing Power Parity (RPPP).
    
    Parameters:
    exchange_rate (float): The current exchange rate (Home currency per Foreign currency)
    inflation_norway (float): The inflation rate in the home country (as a percentage)
    inflation_usa (float): The inflation rate in the foreign country (as a percentage)
    
    Returns:
    float: The new exchange rate as predicted by RPPP.
    """   
    # Calculate the new exchange rate using RPPP formula
    rppp= exchange_rate * ((1 + inflation_norway) / (1 + inflation_usa))
    
    return rppp

# Calculate the new exchange rate
rppp = calculate_rppp_exchange_rate(exchange_rate, inflation_norway, inflation_usa)

print(f"The new exchange rate estimated by RPPP is: {rppp:.4f} NOK/USD")
```

# Porfolio assignment: 6 equally weighted portfolios

```python
def create_portfolios(df3, total_portfolios=6):

    # Initialize a dictionary to store portfolio assignments with the date 
    portfolio_assignments = {'Date': df3['Date'].tolist()}
    # Add portfolio keys to the dictionary
    for i in range(total_portfolios):
        portfolio_assignments[f'Portfolio {i+1}'] = []

    for index, currency_row in df3.iterrows(): 
        # Rank the currencies in the row, excluding NaN values and the 'Date' column
        ranked_currencies = currency_row.dropna().drop('Date').rank(method='min', ascending=True)
        # If there are not enough currencies for the portfolios, add empty lists for this date
        if ranked_currencies.empty or len(ranked_currencies) < total_portfolios:
            for i in range(total_portfolios):
                portfolio_assignments[f'Portfolio {i+1}'].append([])
            continue  # Then countinue

        # Determine the rank boundaries for dividing currencies into portfolios
        rank_boundaries = np.percentile(ranked_currencies.values, np.linspace(0, 100, total_portfolios + 1))

       # Assign each currency to a portfolio based on its rank
        for portfolio_num in range(total_portfolios):
            # Create a condition to check if a currency's rank falls within a portfolio's boundary
            # For the first portfolio (Portfolio_1), the condition includes currencies with ranks 
            # up to the first boundary (inclusive)
            if portfolio_num == 0:
                inclusion_condition = ranked_currencies <= rank_boundaries[portfolio_num + 1]
                 # For other portfolios, include currencies within specific rank ranges
            else:
                inclusion_condition = (ranked_currencies > rank_boundaries[portfolio_num]) & \
                                      (ranked_currencies <= rank_boundaries[portfolio_num + 1])
                # Get the list of currencies that fit the criteria for this portfolio
            assigned_currencies = ranked_currencies[inclusion_condition].index.tolist()
            # Add these currencies to the current portfolio
            portfolio_assignments[f'Portfolio {portfolio_num+1}'].append(assigned_currencies)

    # Convert the dictionary to a DataFrame
    portfolio_df = pd.DataFrame(portfolio_assignments)
    
    return portfolio_df

# Print out the portfolio assignments which include the list of currencies in each portfolio
portfolio_data = create_portfolios(df3, total_portfolios=6)
portfolio_data
```

```python
# Checking the frequency of which currency appear the most in each portfolio
def find_most_frequent_currencies(portfolio_data):
    # Initialize counters for each portfolio
    counter_ptf1 = Counter()
    counter_ptf2 = Counter()
    counter_ptf3 = Counter()
    counter_ptf4 = Counter()
    counter_ptf5 = Counter()
    counter_ptf6 = Counter()

    # Iterate through the DataFrame and count the currencies
    for index, row in portfolio_data.iterrows():
        counter_ptf1.update(row['Portfolio 1'])
        counter_ptf2.update(row['Portfolio 2'])
        counter_ptf3.update(row['Portfolio 3'])
        counter_ptf4.update(row['Portfolio 4'])
        counter_ptf5.update(row['Portfolio 5'])
        counter_ptf6.update(row['Portfolio 6'])

    # Find the most common currency in each portfolio
    most_common_ptf1 = counter_ptf1.most_common(1)[0][0] if counter_ptf1 else None
    most_common_ptf2 = counter_ptf2.most_common(1)[0][0] if counter_ptf2 else None
    most_common_ptf3 = counter_ptf3.most_common(1)[0][0] if counter_ptf3 else None
    most_common_ptf4 = counter_ptf4.most_common(1)[0][0] if counter_ptf4 else None
    most_common_ptf5 = counter_ptf5.most_common(1)[0][0] if counter_ptf4 else None
    most_common_ptf6 = counter_ptf6.most_common(1)[0][0] if counter_ptf6 else None

    return most_common_ptf1,most_common_ptf2,most_common_ptf3,most_common_ptf4,most_common_ptf5, most_common_ptf6

# Call the function with  DataFrame
most_common_currency_ptf1,most_common_currency_ptf2,most_common_currency_ptf3,most_common_currency_ptf4,most_common_currency_ptf5, most_common_currency_ptf6 = find_most_frequent_currencies(portfolio_data)

print("Most frequent currency in Portfolio 1:", most_common_currency_ptf1)
print("Most frequent currency in Portfolio 2:", most_common_currency_ptf2)
print("Most frequent currency in Portfolio 3:", most_common_currency_ptf3)
print("Most frequent currency in Portfolio 4:", most_common_currency_ptf4)
print("Most frequent currency in Portfolio 5:", most_common_currency_ptf5)
print("Most frequent currency in Portfolio 6:", most_common_currency_ptf6)
```

## Average forward discounts

```python
# Calculate the average forward discounts

def calculate_average_forward_discounts(portfolio_data, df3, total_portfolios=6):

    # Initialize a DataFrame to store the average forward discounts for each date
    averages_per_date = pd.DataFrame(index=portfolio_data.index)

    # Iterate through each portfolio
    for i in range(1, total_portfolios + 1):
        portfolio_name = f'Portfolio {i}'
        # Calculate the average forward discount for each portfolio for each date
        averages_per_date[portfolio_name] = portfolio_data.apply(
            lambda row: df3.loc[df3['Date'] == row['Date'], row[portfolio_name]].mean(axis=1).iloc[0] 
                       if isinstance(row[portfolio_name], list) and row[portfolio_name] else np.nan, 
            axis=1
        )

    # Calculate the overall average for each portfolio across all dates
    overall_averages_forward_discount = averages_per_date.mean()

    # Annualize the overall average for each portfolio across all dates
    annualize_overall_averages = overall_averages_forward_discount*12

    return averages_per_date, overall_averages_forward_discount,annualize_overall_averages

# Assuming ptf_df is your portfolio DataFrame and df3 is the forward discount DataFrame
averages_per_date, overall_averages_forward_discount, annualize_overall_averages = calculate_average_forward_discounts(portfolio_data, df3)


# This gives you the overall annual average forward discount for each portfolio across all dates
print('Annually average forward discount: ')
annualize_overall_averages
```

```python
# Calculate the average log excess return, standard deviation and sharp ratio

def logex_stats(ptf_shift, df5, total_portfolios=6):

    # Initialize a DataFrame to store the average log execess  and standard deviation for each date
    averages_logex_date = pd.DataFrame(index=ptf_shift.index)
    std_logex_date = pd.DataFrame(index=ptf_shift.index)

    #Set the decimal up to 9 
    pd.set_option('display.float_format', '{:.9f}'.format)
    

    # Iterate through each portfolio
    for i in range(1, total_portfolios + 1):
        portfolio_name = f'Portfolio {i}' #portfolio name 
        # Calculate the average log execess for each portfolio for each date
        averages_logex_date[portfolio_name] = ptf_shift.apply(
            lambda row: df5.loc[df5['Date'] == row['Date'], row[portfolio_name]].mean(axis=1).iloc[0] 
                       if isinstance(row[portfolio_name], list) and row[portfolio_name] else np.nan, 
            axis=1
        )
     # Iterate through each portfolio
    for i in range(1, total_portfolios + 1):
        portfolio_name = f'Portfolio {i}'  #portfolio name 
        # Calculate the standard deviation of log excess for each portfolio for each date
        std_logex_date[portfolio_name] = ptf_shift.apply(
            lambda row: df5.loc[df5['Date'] == row['Date'], row[portfolio_name]].std(axis=1).iloc[0] 
                       if isinstance(row[portfolio_name], list) and row[portfolio_name] else np.nan, 
            axis=1
        )
    # Calculate the overall average for each portfolio across all dates
    overall_averages_logex = averages_logex_date.mean()

    # Annualize the overall average for each portfolio and display percentage
    annualize_overall_averages_logex = overall_averages_logex*12*100 

    # Calculate the overall standard deviation for each portfolio across all dates
    overall_std_logex = averages_logex_date.std()

    # Annualize the overall standard deviation for each portfolio and display percentage
    annualize_overall_std_logex = overall_std_logex*np.sqrt(12)*100
    
    #Compute the Sharp ratio
    sharp_ratio = annualize_overall_averages_logex/annualize_overall_std_logex

   

    # Creating a new DataFrame to display the results
    results_df = pd.DataFrame(index=[f'Portfolio {i}' for i in range(1, total_portfolios + 1)])
    results_df['Annual Log Return (%)'] = annualize_overall_averages_logex
    results_df['Annual Std Dev (%)'] = annualize_overall_std_logex
    results_df['Sharpe Ratio'] = sharp_ratio

    return averages_logex_date,std_logex_date,overall_averages_logex,annualize_overall_averages_logex,overall_std_logex,annualize_overall_std_logex,sharp_ratio,results_df

averages_logex_date,std_logex_date,overall_averages_logex,annualize_overall_averages_logex,overall_std_logex,annualize_overall_std_logex,sharp_ratio,results_df = logex_stats(ptf_shift, df5)

# Print the DataFrame with annual log return, annual standard return, and Sharpe rati
print(results_df)
```

```python

def logex_diff(ptf_shift, df5):
    # Initialize a Series to store the difference for each date
    diff_logex = pd.Series(index=ptf_shift.index)

    for index, row in ptf_shift.iterrows():
        # Assuming each row of ptf_shift contains the asset names for each portfolio
        last_portfolio_assets = row['Portfolio 6']
        first_portfolio_assets = row['Portfolio 1']

        # Directly compute the difference in log excess returns
        # between the last and first portfolios
        last_portfolio_return = df5.loc[index, last_portfolio_assets]
        first_portfolio_return = df5.loc[index, first_portfolio_assets]

        # Calculate the difference and assign it to the diff_logex series
        diff_logex[index] = last_portfolio_return.mean() - first_portfolio_return.mean()

    return diff_logex

# Assuming ptf_shift and df5 are your dataframes
diff_logexcess = logex_diff(ptf_shift, df5)

# Perform a one-sample t-test
t_stat, p_value = stats.ttest_1samp(diff_logexcess.dropna(), 0)

print("T-statistic:", t_stat)
print("p-value:", p_value)

alpha = 0.05  # Set alpha
if p_value < alpha:
    print(f"Difference is statistically significant (p-value: {p_value})")
else:
    print(f"Difference is not statistically significant (p-value: {p_value})")
```
